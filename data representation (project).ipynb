```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.axis import Axis  
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
x = pd.read_excel (r'D:\BLE_RSSI_dataset\iBeacon_RSSI_Labeled.xlsx')
x.describe()
```

```python
strong_sig1 = x['b3001']>-85
a = x.loc[strong_sig1,'location'].value_counts()
plt.figure(figsize=(20,5))
plt.plot(a, color = 'r')
plt.title("beacon 1") 
plt.xlabel("Location")
plt.ylabel("Num of RSSI readings > -85")
plt.grid()
plt.savefig("D:\\BLE_RSSI_dataset\\be001.jpeg")
```
